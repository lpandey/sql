```python
# using python variables -- can be used with a ":" prefix
country = "Canada"
%sql select * from INTERNATIONAL_STUDENT_TEST_SCORES where country = :country
```

```python
# assign query results to python variables
test_score_distribution = %sql SELECT test_score as "Test Score", count(*) as "Frequency" from INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY test_score;
test_score_distribution
```

```python
# query result into dataframe
df = test_score_distribution.DataFrame()

# plot
plot = seaborn.barplot(x='Test Score', y='Frequency', data=df)
```
